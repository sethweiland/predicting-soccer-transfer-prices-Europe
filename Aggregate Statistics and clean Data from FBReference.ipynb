```python
df_list = []
for csv in stat_csvs:
    df_temp = pd.read_csv(csv)
    df_temp.columns = df_temp.iloc[0,:].values
    df_temp['Player_new'] = df_temp.iloc[:,1]
    df_temp['Player_split'] = df_temp['Player_new'].str.split("\\")
    df_temp['Player_name'] = df_temp['Player_split'].apply(get_name)
    df_temp['Player_under'] = df_temp['Player_name'].str.split(" ").str.join("_")
    str_test = csv
    split_list = str_test.split("_")
    year_one = int(split_list[1])
    year_two = year_one+1
    df_temp['Season_Transfer'] = df_temp['Player_under'] + f" 20{year_one}-{year_two}"
    print(df_temp.columns)
    df_list.append(df_temp)
    
```

```python
#this is because this one csv had irregular columns as compared to the others
df_list[2].columns = df_list[3].columns
```
